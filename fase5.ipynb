```python
#receiver operating characteristic(ROC) curve
#√Årea bajo la curva (AUC): El AUC es una medida del rendimiento general del clasificador. Un AUC de 1 indica una clasificaci√≥n perfecta, mientras que
#un AUC de 0,5 indica una clasificaci√≥n no mejor que aleatoria. En este caso, el AUC es de aproximadamente 0,87, lo que sugiere que el clasificador 
#est√° funcionando bien.


#Forma de la curva: La forma de la curva ROC puede proporcionar informaci√≥n sobre las fortalezas y debilidades del clasificador. Una curva que se 
#eleva abruptamente y se acerca a la esquina superior izquierda indica que el clasificador es capaz de distinguir bien entre positivos y negativos.
#En este caso, la curva tiene una buena forma, lo que indica que el clasificador es generalmente efectivo.
```
